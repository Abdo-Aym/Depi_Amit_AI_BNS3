```python
import pandas as pd
import matplotlib.pyplot as plt
import os
# os.chdir(R"C:\Users\amb\Downloa/ds")
```

```python
dataset = pd.read_csv(r'G:/DEPI/Depi_Amit_AI_BNS3/Sources/Machine_Learning/session2/datadata.txt')
dataset  
```

```python
df = dataset.copy()
```

```python
X = df.iloc[: ,:-1]
y = df.iloc[: , -1]
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression

lin = LinearRegression()
lin.fit(X,y)
```

```python
from sklearn.linear_model import Ridge,Lasso  # reach to zero so we can use it for feature selection
```

# Ridge

```python
l = 10000
Ridge = Ridge(alpha = l)
Ridge.fit(X,y)
```

```python
Ridge_coeff = Ridge.coef_
Ridge_coeff
```

```python
Ridge_intercept = Ridge.intercept_
Ridge_intercept
```

```python
X_test = [1,2,3,4]
y_pred = Ridge.predict(pd.DataFrame(X_test))
y_pred
```

```python
plt.figure(figsize=(12,6) , dpi = 90)
plt.plot(X_test,y_pred , c ='r' , linewidth = 2)
plt.scatter(X,y)
plt.ylim(ymin = 0 , ymax = 9)
plt.xlim(xmin = 0 , xmax = 7)

plt.text(X_test[-1],y_pred[-1] , ' y = '   +    str('%.2f' %Ridge_coeff)  + ' * X +'  +  str('%.2f' %Ridge_intercept)    
                                                   +  '  For \u03BB or  \u03b1 = ' +    str(l)  ,   fontsize = 12)
plt.title('Ridge Regression')
plt.show()
```

# Lasso

```python
l = 2.5
Lasso = Lasso(alpha=l)
Lasso.fit(X,y)
```

```python
Lasso_coeff = Lasso.coef_
Lasso_coeff
```

```python
Lasso_intercept = Lasso.intercept_
Lasso_intercept
```

```python
X_test = [1,2,3,4]
y_pred = Lasso.predict(pd.DataFrame(X_test))
y_pred
```

```python
plt.figure(figsize=(12,6) , dpi = 90)
plt.plot(X_test,y_pred , c ='r' , linewidth = 2)
plt.scatter(X,y)
plt.ylim(ymin = 0 , ymax = 9)
plt.xlim(xmin = 0 , xmax = 7)

plt.text(X_test[-1],y_pred[-1] , ' y = '   +    str('%.2f' %Lasso_coeff)  + ' * X +'  +  str('%.2f' %Lasso_intercept)    
                                                   +  '  For \u03BB or  \u03b1 = ' +    str(l)  ,   fontsize = 12)
plt.title('Lasso Regression')
plt.show()
```
